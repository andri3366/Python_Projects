```python
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

df = pd.read_csv("../data/Students Social Media Addiction.csv")
```

```python
df_clean = df.dropna(subset=["Affects_Academic_Performance", "Avg_Daily_Usage_Hours"])
```

```python
academic_usage = df_clean.groupby(["Academic_Level", "Affects_Academic_Performance",])["Avg_Daily_Usage_Hours"].mean().reset_index().rename(columns={"Avg_Daily_Usage_Hours" : "Avg_Hours"})
academic_usage
```

```python
plt.figure(figsize=(8,6))
sns.barplot(
    x="Academic_Level",
    y="Avg_Hours",
    hue="Affects_Academic_Performance",
    data=academic_usage
)

plt.title("Impact of Social Media on Academic Performance")
plt.xlabel("Academic Level")
plt.ylabel("Average Daily Usage (hours)")
plt.legend(title="Affects Academic Performance", bbox_to_anchor=(1.05, 1), loc='best')
plt.tight_layout()
plt.show()
```
