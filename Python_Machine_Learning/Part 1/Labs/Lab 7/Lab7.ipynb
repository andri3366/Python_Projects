## Data Cleaning and Feature Engineering

```python
df.sample(3)
```

```python
duplicates = df.duplicated()
duplicates[duplicates==True]
df.iloc[[]]
```

```python
df.dtypes
```

```python
df2 = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
df2.dtypes
```

```python
df2['Age'] = df2['Age'].fillna(df2['Age'].median())
```

```python
df2['Embarked'] = df2['Embarked'].fillna(df2['Embarked'].mode()[0])
```

```python
df2 = pd.get_dummies(df2, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
df2.dtypes
```

```python
df2.sample()
```

```python
df2.to_csv('clean_titanic.csv', index=None)
new = pd.read_csv('clean_titanic.csv')
new.head()
```
