## Task 4 Models

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('clean_titanic.csv')
df.sample(3)
```

```python
df.shape
```

```python
x = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
def model_evaluation(test_size):
    x_train, x_test, y_train, y_test = train_test_split(
        x,
        y, 
        test_size=test_size, 
        random_state=42)
    
    print("TEST model shape: ", x_train.shape, y_train.shape, x_test.shape, y_test.shape)
    print("------------")
    models = {
        "Logistic Regression": LogisticRegression(max_iter=1000),
        "Decision Tree": DecisionTreeClassifier(),
        "Random Forest": RandomForestClassifier()
    }

    results = {}

    for name, model in models.items():
        print(name)
        model.fit(x_train, y_train)

        train_pred = model.predict(x_train)
        # print("TEST train prediciton: ",train_pred)
        test_pred = model.predict(x_test)
        # print("TEST test prediciton: ",test_pred)
        # print("------------")

        train_acc = accuracy_score(y_train, train_pred)
        print("TEST train accuracy: ",train_acc)
        test_acc = accuracy_score(y_test, test_pred)
        print("TEST test accuracy: ",test_acc)

        cm = confusion_matrix(y_test, test_pred)

        print("Confusion Matrix:")
        print(cm)
        print("------------")

        cv_scores = cross_val_score(model, x, y, cv=5)

        print("Cross Validation Score, k=5")
        print(cv_scores)
        print("Average CV")
        print(cv_scores.mean())
        print("------------")
        results[name] = test_acc
    
    return results
```

```python
print("80/20")
model_evaluation(0.2)
```

```python
print("70/30")
model_evaluation(0.3)
```

```python
print("85:15")
model_evaluation(0.15)
```

## Task 5 Tune Hyperparameters

- n_estimators: number of decision trees in the forest
- max_depth: longest the path from the root to leaf in any tree
- max_features: determines the maximum number of features the algorithms considers when looking for the best split in the nodes

```python
rf_tune = []
def tune_random_forest(n_estimators, max_depth=None, max_features=None):

    model = RandomForestClassifier(
        n_estimators = n_estimators,
        max_depth = max_depth,
        max_features = max_features,
        random_state = 42
    )

    cv_scores = cross_val_score(model, x, y, cv=5)

    print(f"Hyper Parameter Tuning: {n_estimators}, {max_depth}, {max_features}")
    print("Cross Validation Score, k=5")
    print(cv_scores)

    avg_score = cv_scores.mean()
    print("Average CV")
    print(avg_score)
    print("------------")

    rf_tune.append({
        "n_estimators": n_estimators,
        "max_depth": max_depth,
        "max_features": max_features,
        "avg_score": avg_score
    })
```

```python
tune_random_forest(100)
tune_random_forest(100, 6)
tune_random_forest(100, 6, 'sqrt')
tune_random_forest(200, 6, 'sqrt')
tune_random_forest(100, 6, 'log2')
tune_random_forest(200, 6, 'log2')
tune_random_forest(200)
tune_random_forest(200, 8)
tune_random_forest(200, 8, 'sqrt')
tune_random_forest(100, 8, 'sqrt')
tune_random_forest(200, 8, 'log2')
tune_random_forest(100, 8, 'log2')
```

```python
best_tune = max(rf_tune, key=lambda x: x["avg_score"])

print("Best Hyperparameter Combination:")
print(best_tune)
```
