```python
# Regression Model
x = df.drop('price', axis=1)
y = df['price']
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def model_evaluation(test_size):
    x_train, x_test, y_train, y_test = train_test_split(
        x,
        y, 
        test_size=test_size, 
        random_state=42)
    
    print("TEST model shape: ", x_train.shape, y_train.shape, x_test.shape, y_test.shape)
    print("------------")
    model = LinearRegression()
    lrmodel = model.fit(x_train, y_train)
    lr_coef = lrmodel.coef_
    lr_intercept = lrmodel.intercept_

    train_pred = lrmodel.predict(x_train)
    print("TEST train prediciton: ",train_pred)
    test_pred = lrmodel.predict(x_test)
    print("TEST test prediciton: ",test_pred)
    print("------------")

    # Mae
    train_mae = mean_absolute_error(train_pred, y_train)
    test_mae = mean_absolute_error(test_pred, y_test)

    # RMSE
    train_mse = mean_squared_error(train_pred, y_train)
    train_rmse = np.sqrt(train_mse)

    test_mse = mean_squared_error(test_pred, y_test)
    test_rmse = np.sqrt(test_mse)

    print(f"Test split: {test_size}")
    print("Train MAE: ", train_mae)
    print("Test MAE: ", test_mae)
    print("Train RMSE: ", train_rmse)
    print("Test MAE: ", test_rmse)
    print("------------")
    print("Coef: ", lr_coef)
    print("Intercept: ", lr_intercept)

    coef_df = pd.DataFrame({
        "Feautre" : x.columns, 
        "Coefficient" : lr_coef
    })

    top_pos = coef_df.sort_values(by="Coefficient", ascending=False).head(3)
    top_neg = coef_df.sort_values(by="Coefficient").head(3)

    print("Top 3 Positive: ")
    print(top_pos)
    print("Top 3 Negative: ")
    print(top_neg)
    
    return model, test_rmse
```

```python
model_80, rmse_80 = model_evaluation(0.2)
```

```python
model_70, rmse_70 = model_evaluation(0.3)
```

```python
model_85, rmse_85 = model_evaluation(0.15)
```
