# Task 3

### Column Conversion

```python
df.dtypes
```

```python
df["hasYard"] = df["hasYard"].map({"Yes":1, "No": 0})
```

```python
df["hasPool"] = df["hasPool"].map({"Yes":1, "No": 0})
```

```python
df["isNewBuilt"] = df["isNewBuilt"].map({"Yes":1, "No": 0})
```

```python
df["hasStormProtector"] = df["hasStormProtector"].map({"Yes":1, "No": 0})
```

```python
df.dtypes
```

### Feature Engineering

```python
df['Popular'] = ((df.hasPool == 1) & (df.isNewBuilt == 1)).astype(int)
df.sample(6)
```

```python
df.Popular.value_counts()
```

### Interaction Features

```python
df["made"].max()
```

```python
df["houseAge"] = 2026 - df["made"]
df.head()
```

```python
df = df.drop("made", axis=1)
```

```python
df.sample()
```

```python
df["numNonGuestRooms"] = df.numberOfRooms - df.hasGuestRoom
df.sample()
```

```python
df["grossLivingArea"] = df.squareMeters - (df.basement + df.attic + df.garage)
df.sample()
```

```python
df = df[df.grossLivingArea >= 0]
```
