# Task 1
- Use NumPy as np to crete a 2D array (5x5) with random integers between 10 and 50
- Determine the mean, median, and standard deviation along with the max and min elements and their indices

```python
print("Student Name: {sn} \nStudent Number: {num}".format(sn="Andrianna Wardill", num="041067235"))
```

```python
import numpy as np

low = 10
high = 50
shape = (5,5)
```

```python
rand_array = np.random.randint(low, high, size=shape)
```

```python
print(rand_array)
```

```python
data = np.array(rand_array)
# data
```

```python
mean = np.mean(data)
median = np.median(data)
std = np.std(data)
min_val = np.min(data)
min_index = tuple(int(i) for i in np.unravel_index(data.argmin(), data.shape))
max_val = np.max(data)
max_index = tuple(int(i) for i in np.unravel_index(data.argmax(), data.shape))
```

```python
print(f"Mean: {mean}")
print(f"Median: {median}")
print("Std: %.2f" % std)
print(f"Min Value: {min_val}")
print(f"Min Index: {min_index}")
print(f"Max Value: {max_val}")
print(f"Max Index: {max_index}")
```

# Task 2
- Use scipy.stats and scipy.integrate to create a dataset of 100 random numbers from a normal distribution with a mean of 50 and std of 10
- Comput the mean, median, mode, variance, skewness, and kurtosis

```python
from scipy import stats, integrate
loc = 50
scale = 10
```

```python
data = stats.norm.rvs(loc=loc, scale=scale, size=100)
```

```python
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True)[0][0]
var = np.var(data)
skew = stats.skew(data)
kurtosis = stats.kurtosis(data)
```

```python
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
print(f"Variance: {var:.2f}")
print(f"Skew: {skew:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

```python
pdf = lambda x: stats.norm.pdf(x, loc, scale)
```

```python
lower = loc - 2 * scale
upper = loc + 2 * scale
```

```python
result, err = integrate.quad(pdf, lower, upper)
print(f"Probability between {lower} and {upper}: {result:.2f}")
```
