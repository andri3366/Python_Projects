```python
# Task Requirements - 4
import logging
from functools import reduce

logging.shutdown()

logging.basicConfig(
    filename="error_AW.log",
    format='%(asctime)s - %(levelname)s - %(message)s',
    level=logging.ERROR
)

def error_handling(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)

        except(ValueError, TypeError, IndexError) as e:
            logging.error(f"{func.__name__}: {e}")
            # print(f"Error ({func.__name__}) : {e}")
            return None
    
        
    return wrapper
```

### Task requirement 5 completed below

```python
# Task Requirements - 5

@error_handling
def gen_data(start, stop):

    if not isinstance(start, int) or not isinstance(stop, int):
        raise TypeError("Must be of type int")
    if start > stop:
        raise ValueError("Start cannot be greater than stop")
    
    # added a nested function for error handling to work
    def generator():
        for i in range(start, stop + 1):
            yield i
    
    return generator()
    # for i in range(start, stop + 1):
    #     yield i


@error_handling
def map_data(dataset, ascii_sum):
    return list(map(lambda x: x**2 if ascii_sum % 2 == 0 else x**3, dataset))

@error_handling
def filter_data(dataset, threshold):
    return list(filter(lambda x: x > threshold, dataset))

@error_handling
def reduce_data(dataset, num):
    return reduce(lambda a,b: a + b if num % 2 == 0 else a * b, dataset)
```
