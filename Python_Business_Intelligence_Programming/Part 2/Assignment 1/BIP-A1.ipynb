```python
# Task Requirements - 4
import logging
from functools import reduce

logging.shutdown()

logging.basicConfig(
    filename="error_AW.log",
    format='%(asctime)s - %(levelname)s - %(message)s',
    level=logging.ERROR
)

def error_handling(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)

        except(ValueError, TypeError, IndexError) as e:
            logging.error(f"{func.__name__}: {e}")
            # print(f"Error ({func.__name__}) : {e}")
            return None
    
        
    return wrapper
```

```python
# Task Requirement - 6

num = "041067235"

ascii_val = [ord(char) for char in num]
print(ascii_val)

# Start value
start = sum((ascii_val[0], ascii_val[1])) % 50
print(f"Start = {start}")

# Stop value
stop = start + (sum((ascii_val[-1], ascii_val[-2])) % 30 + 20)
print(f"Stop = {stop}")

# Filter threshold
threshold = (sum(ascii_val) % 25) + 10
print(f"Threshold = {threshold}")

dataset = list(gen_data(start, stop))
print(dataset)

ascii_sum = sum(ascii_val)
mapped = map_data(dataset, ascii_sum)

if mapped is not None:
    print(f"Mapped Data: {mapped}")
    filtered = filter_data(mapped, threshold)

    if filtered is not None:

        print(f"Filtered Data: {filtered}")

        result = reduce_data([filtered], len(num))

        if result is not None:

            print(f"Reduced Data: {result}")

print("\nLog file (error_AW.log)")
try:
    with open("error_AW.log", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("No log found")
```
