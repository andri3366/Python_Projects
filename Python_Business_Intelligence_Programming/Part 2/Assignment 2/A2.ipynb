```python
## added logging for debugging after demo
import logging
from functools import reduce

logging.shutdown()

logging.basicConfig(
    filename="error_AW.log",
    format='%(asctime)s - %(levelname)s - %(message)s',
    level=logging.ERROR
)

def error_handling(func):
    def wrapper(*args, **kwargs):
        
        try:
            return func(*args, **kwargs)
        except(ValueError, ModuleNotFoundError) as e:
            logging.error(f"{func.__name__}: {e}")
    
    return wrapper

import importlib

# added decorator
@error_handling
def calculate_area():
    radius = float(input("Enter the radius: "))
    return module.pi * radius * radius

@error_handling
def get_module():

    module_name = input("Enter a library name: ")
    return importlib.import_module(module_name)
       
# main function for scripting 
if __name__ == "__main__":
    module = get_module()
    
    if module.__name__ == "math":
        print("Area:", calculate_area())
    else:
        print("No operations")
```
