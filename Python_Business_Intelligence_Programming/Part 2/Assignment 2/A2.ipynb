### Task 3
Python scripts that reads data from a txt file and counts the number of words and lines then write the results to a new file.

```python
word_count = 0
line_count = 0

with open("data.txt", 'r') as file:
    for line in file:
        words = line.split()
        word_count += len(words)
        line_count += 1
        
print(f"Word Count: {word_count}")
print(f"Line count: {line_count}")

with open("results.txt", 'w') as file:
    file.write(f"Word Count of data.txt: {word_count}\n")
    file.write(f"Line count of data.txt: {line_count}")
```
