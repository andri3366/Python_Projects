## Question 1 - Encapsulation
Scenario: A Bank Account system must protect balance

```python
class BankAccount():
    def __init__(self, balance=0):
        self.__balance = balance
    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited: {amount}")
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else :
            print("Withdrawing more than account balance")
    def get_balance(self):
        print(f"Balance: {self.__balance}")        
```

```python
a = BankAccount(1000)
```

```python
a.deposit(200)
a.get_balance()
a.withdraw(150)
a.get_balance()
```

## Question 2 - Inheritance
Scenario: School Management System

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")
```

```python
p1 = Person("Alice", 20)
s1 = Student("Ana", 22, 100)
```

```python
p1.display_info()
s1.display_info()
```

## Question 3 - Polymorphism
Scenario: Different animal sounds

```python
class Animal:
    def sound(self):
        print("Animal Sound!")

class Dog(Animal):
    def sound(self):
        print("Woof!")

class Cat(Animal):
    def sound(self):
        print("Meow!")
```

```python
dog = Dog()
cat = Cat()
```

```python
dog.sound()
cat.sound()
```

## Question 4 - Protected Members
Scenario: Compnay employee data

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self._salary = salary

class Manager(Employee):
    def __init__(self, name, salary):
        super().__init__(name, salary)
    def display(self):
        print(f"Name: {self.name}")
        print(f"Salary: {self._salary}")
```

```python
m1 = Manager("Bob", 60000)
```

```python
m1.display()
```

## Question 5 - Combined OOP
Scenario: Vehical Rental System

```python
class Vehicle:
    def __init__(self, price):
        self.price = price
    def rent_price(self):
        print(f"Vehicle Price: {self.price}")

class Car(Vehicle):
    def __init__(self, price):
        super().__init__(price)
    def rent_price(self):
        print(f"Car Price: {self.price}")

class Bike(Vehicle):
    def __init__(self, price):
        super().__init__(price)
    def rent_price(self):
        print(f"Bike Price: {self.price}")
```

```python
v1 = Vehicle(10000)
c1 = Car(8000)
b1 = Bike(200)
```

```python
v1.rent_price()
c1.rent_price()
b1.rent_price()
```
