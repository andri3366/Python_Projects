### Question : 2 Collections 

Imagine you are building a program to manage a bookstore's inventory. You need to store information about each book, including its title, author, genre, and price.  Provide an example of how you would use this collection to store the details of at least three different books.



```python
from collections import defaultdict
```

```python
inventory = defaultdict(list)
```

```python
inventory["Classic"].append({
    "Title": "Great Gatsby",
    "Author" : "Fitzgerald",
    "Price" : 23.45
})

inventory["Fiction"].append({
    "Title": "1984",
    "Author" : "George Orwell",
    "Price" : 19.84
})

inventory["Fantasy"].append({
    "Title": "Harry Potter Deathly Hollows Part 1",
    "Author" : "J.K. Rowling",
    "Price" : 21.50
})
```

```python
for genre, books in inventory.items():
    print(genre)
    for book in books:
        print(f"{book["Title"]} - {book["Author"]} - {book["Price"]}")
```

```python
birthday = datetime(2004, 8, 22)
today = datetime.today()

next_birthday = birthday.replace(year=today.year)
```

```python
if next_birthday < today:
    next_birthday = next_birthday.replace(year=today.year + 1)
days = (next_birthday - today).days
```
