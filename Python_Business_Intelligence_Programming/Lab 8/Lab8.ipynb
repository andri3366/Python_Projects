## Question 1: School Management System (Inheritance)
Scenario: A school system stores information about people in the school. Both Students and Teachers share some common features such as name and ID

```python
class Person:
    def __init__(self, name, id):
        self.name = name
        self.id = id
    def info(self):
        print(f"Name: {self.name}")
        print(f"Id: {self.id}")

class Student(Person):
    def show_info(self):
        self.info()
        print(f"Role: Student")

class Teacher(Person):
    def show_info(self):
        self.info()
        print(f"Role: Teacher")
```

```python
s1 = Student("Bob", 100)
t1 = Teacher("Ana", 200)
```

```python
s1.show_info()
```

```python
t1.show_info()
```

## Question 2: Employee Salary System (Method Overriding)
Scenario: In a company, different employees calculate their salary differently. Managers receive additional bonuses compared to regular employees. 

```python
class Employee():
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    def details(self):
        print(f"Name: {self.name}")
        print(f"Salary: {self.salary}")

class Manager(Employee):
    def details(self):
        super().details()
        bonus = int(self.salary * 0.10)
        print(f"Bonus: {bonus}")
        print(f"Total Salary: {self.salary + bonus}")
```

```python
e1 = Employee("Bob", 22000)
m1 = Manager("Ana", 25000)
```

```python
e1.details()
```

```python
m1.details()
```

## Question 3: Payment System (Ploymorphism)
Scenario: An online store allows customers to pay using different methods such as Credit Card, PayPal, or DebitCard

```python
class CreditCard:
    def process_payment(self):
        print("Credit Card")

class PayPal:
    def process_payment(self):
        print("PayPal")

class DebitCard:
    def process_payment(self):
        print("Debit Card")

def make_payment(payment_method):
    payment_method.process_payment()
```

```python
p1 = CreditCard()
p2 = PayPal()
p3 = DebitCard()
```

```python
make_payment(p1)
make_payment(p2)
make_payment(p3)
```
